## Analise Total de Bairros por Acidentes, Arrecadações,  Buracos ,Velocidade, Temperatura 

```python
bairros_total_acidentes = governo_estadual.groupby(["Bairros"]).sum()["Acidentes"].rename("Total de Acidentes") 
bairros_total_arrecadacoes = governo_estadual.groupby(["Bairros"]).sum()["Arrecadado"].rename("Total de Arrecadações")
bairros_total_buracos = governo_estadual.groupby(["Bairros"]).sum()["Buracos"].rename("Total de Buracos")
bairros_total_velocidade = governo_estadual.groupby(["Bairros"]).sum()["Velocidade"].rename("Total de Velocidade")
bairros_total_temperatura = governo_estadual.groupby(["Bairros"]).sum()["Temperatura"].rename("Total de Temperatura")

analise_bairros_total = pd.DataFrame({"Total de Acidentes" : bairros_total_acidentes, 
                                "Total de Arrecadações" : bairros_total_arrecadacoes, 
                                "Total de Buracos" : bairros_total_buracos, 
                                "Total de Temperatura" : bairros_total_temperatura, 
                                "Total de Velocidade" : bairros_total_velocidade}) 

analise_bairros_total ["Total de Acidentes"] = analise_bairros_total["Total de Acidentes"].map("{:,.2f}".format)
analise_bairros_total ["Total de Arrecadações"] = analise_bairros_total["Total de Arrecadações"].map(" R$ {:,.2f}".format)
analise_bairros_total ["Total de Buracos"] = analise_bairros_total["Total de Buracos"].map("{:,.2f}".format)
analise_bairros_total ["Total de Temperatura"] = analise_bairros_total["Total de Temperatura"].map("{:,.2f} °C".format)
analise_bairros_total ["Total de Velocidade"] = analise_bairros_total["Total de Velocidade"].map("{:,.2f}".format)


analise_bairros_total
```

## Moda de Bairros por Acidentes, Arrecadações, Buracos, Temperatura e Velocidade

```python
moda_analise = pd.DataFrame() 

for b in list(governo_estadual["Bairros"].unique()): 
    
    ac = governo_estadual[(governo_estadual["Bairros"] == b)]["Acidentes"].mode()
    a = governo_estadual[(governo_estadual["Bairros"] == b)]["Arrecadado"].mode()
    bb = governo_estadual[(governo_estadual["Bairros"] == b)]["Buracos"].mode()
    t = governo_estadual[(governo_estadual["Bairros"] == b)]["Temperatura"].mode()
    v = governo_estadual[(governo_estadual["Bairros"] == b)]["Velocidade"].mode()
    moda_analise = moda_analise.append(pd.DataFrame({'Bairros': b, 'Moda de Acidentes':ac, 'Moda Arrecadadas': a, 
                                                     'Moda de Buracos':bb, 'Moda de Temperatura' : t, 'Moda de Velocidade' : v}, index=[0]))
        
moda_analise = moda_analise.reset_index() 
del moda_analise["index"] 

moda_analise["Moda de Acidentes"].fillna(0, inplace=True) 
moda_analise["Moda de Acidentes"].isnull().value_counts()
moda_analise["Moda de Acidentes"] = moda_analise["Moda de Acidentes"].map(" {:,.2f}".format)

moda_analise["Moda Arrecadadas"].fillna(0, inplace=True) 
moda_analise["Moda Arrecadadas"].isnull().value_counts()
moda_analise["Moda Arrecadadas"] = moda_analise["Moda Arrecadadas"].map("R$ {:,.2f}".format)

moda_analise["Moda de Buracos"].fillna(0, inplace=True) 
moda_analise["Moda de Buracos"].isnull().value_counts()
moda_analise["Moda de Buracos"] = moda_analise["Moda de Buracos"].astype(int) 

moda_analise["Moda de Temperatura"].fillna(0, inplace=True) 
moda_analise["Moda de Temperatura"].isnull().value_counts()
moda_analise["Moda de Temperatura"] = moda_analise["Moda de Temperatura"].map("{:,.2f} °C".format)

moda_analise["Moda de Velocidade"].fillna(0, inplace=True) 
moda_analise["Moda de Velocidade"].isnull().value_counts()
moda_analise["Moda de Velocidade"] = moda_analise["Moda de Velocidade"].map("{:,.2f}".format)

moda_analise
```

```python
variancia_acidentes = governo_estadual.groupby(["Meses"]).var()["Acidentes"].rename("Variância de Acidentes")
variancia_arrecadados = governo_estadual.groupby(["Meses"]).var()["Arrecadado"].rename("Variância de Arrecadações")
variancia_buracos = governo_estadual.groupby(["Meses"]).var()["Buracos"].rename("Variância de Buracos")
variancia_temperatura = governo_estadual.groupby(["Meses"]).var()["Temperatura"].rename("Variância de Temperatura")
variancia_velocidade = governo_estadual.groupby(["Meses"]).var()["Velocidade"].rename("Variância de Velocidade") 


analise_variancia_meses = pd.DataFrame({"Variância de Acidentes" : variancia_acidentes, 
                                    "Variância de Arrecadações" : variancia_arrecadados, 
                                    "Variância de Buracos" : variancia_buracos, 
                                    "Variância de Temperatura" : variancia_temperatura, 
                                    "Variância de Velocidade" : variancia_velocidade}) 

analise_variancia_meses["Variância de Acidentes"] = analise_variancia_meses["Variância de Acidentes"].map("{:,.2f}".format)
analise_variancia_meses["Variância de Arrecadações"] = analise_variancia_meses["Variância de Arrecadações"].map("R$ {:,.2f}".format) 
analise_variancia_meses["Variância de Buracos"] = analise_variancia_meses["Variância de Buracos"].map("{:,.2f}".format)
analise_variancia_meses["Variância de Temperatura"] = analise_variancia_meses["Variância de Temperatura"].map("{:,.2f} °C".format)
analise_variancia_meses["Variância de Velocidade"] = analise_variancia_meses["Variância de Velocidade"].map("{:,.2f}".format)
analise_variancia_meses
```

## Moda de Meses por Acidentes, Arrecadações, Buracos, Temperatura e Velocidade

```python
moda_analise1 = pd.DataFrame() 

for m in list(governo_estadual["Meses"].unique()): 
    
    ac = governo_estadual[(governo_estadual["Meses"] == m)]["Acidentes"].mode()
    a = governo_estadual[(governo_estadual["Meses"] == m)]["Arrecadado"].mode()
    bb = governo_estadual[(governo_estadual["Meses"] == m)]["Buracos"].mode()
    t = governo_estadual[(governo_estadual["Meses"] == m)]["Temperatura"].mode()
    v = governo_estadual[(governo_estadual["Meses"] == m)]["Velocidade"].mode()
    moda_analise1 = moda_analise1.append(pd.DataFrame({'Meses': m, 'Moda de Acidentes':ac, 'Moda Arrecadadas': a, 
                                                     'Moda de Buracos':bb, 'Moda de Temperatura' : t, 
                                                     'Moda de Velocidade' : v}, index=[0]))
        
moda_analise1 = moda_analise1.reset_index() 
del moda_analise1["index"] 

moda_analise1["Moda de Acidentes"].fillna(0, inplace=True) 
moda_analise1["Moda de Acidentes"].isnull().value_counts()
moda_analise1["Moda de Acidentes"] = moda_analise1["Moda de Acidentes"].map(" {:,.2f}".format)

moda_analise1["Moda Arrecadadas"].fillna(0, inplace=True) 
moda_analise1["Moda Arrecadadas"].isnull().value_counts()
moda_analise1["Moda Arrecadadas"] = moda_analise1["Moda Arrecadadas"].map("R$ {:,.2f}".format)

moda_analise1["Moda de Buracos"].fillna(0, inplace=True) 
moda_analise1["Moda de Buracos"].isnull().value_counts()
moda_analise1["Moda de Buracos"] = moda_analise1["Moda de Buracos"].astype(int) 

moda_analise1["Moda de Temperatura"].fillna(0, inplace=True) 
moda_analise1["Moda de Temperatura"].isnull().value_counts()
moda_analise1["Moda de Temperatura"] = moda_analise1["Moda de Temperatura"].map("{:,.2f} °C".format)

moda_analise1["Moda de Velocidade"].fillna(0, inplace=True) 
moda_analise1["Moda de Velocidade"].isnull().value_counts()
moda_analise1["Moda de Velocidade"] = moda_analise1["Moda de Velocidade"].map("{:,.2f}".format)

moda_analise1
```
